## <font color='red'>  --- TODO 3: 分析哪种类型的游戏最赚钱？ ---</font>


```python
# 任务：1. 使用 .groupby() 按 '类型' 分组。
#       2. 计算每个游戏类型的 '全球销量(百万)' 的总和 (sum)。
#       3. 将结果存储在名为 `sales_by_genre` 的变量中，并打印出来。
```

<font color='red'>在下面的cell里写代码</font>

```python
print("\n--- 各游戏类型的全球总销量 ---")
#sales_by_genre = # 在这里写代码 df_games.groupby()[''].sum()
sales_by_genre = df_games.groupby('类型')['全球销量(百万)'].sum()
sales_by_genre
```

## <font color='red'> --- TODO 4: 可视化游戏类型销量 --- </font>


```python
# 任务：调用 `sales_by_genre` 变量的 .plot() 方法，绘制一个条形图(kind='bar')来展示不同游戏类型的总销量。
#       并使用 plt.title() 等函数为图表添加合适的标题和坐标轴标签。  
```

<font color='red'>在下面的cell里写代码</font>

```python
print("\n--- 绘制作物产量对比图 ---")
sales_by_genre.sort_values(ascending=False).plot(kind='bar', figsize=(12, 6))# 在这里写代码
plt.title('不同游戏类型的全球总销量对比')
plt.xlabel('游戏类型')
plt.ylabel('全球总销量 (百万)')
plt.show()
```

## <font color='red'>--- 【挑战1】计算并可视化历史上总销量最高的5个游戏平台是哪些？ ---</font>

```python
print("\n\n--- 【挑战1】总销量最高的Top 5游戏平台 ---")
    
# 步骤1: 按 '平台' 分组，并计算 '全球销量(百万)' 的总和  对结果进行降序排序，并用 .head() 选取前5个
# 任务: 1. 按 '平台' 分组, 2. 计算 '全球销量(百万)' 的总和, 3. 降序排序, 4. 选取前5名
top_5_platforms = df_games.groupby('平台')['全球销量(百万)'].sum().sort_values(ascending=False).head(5)
top_5_platforms
```

```python
print("历史上总销量最高的5个游戏平台是：\n", top_5_platforms)
    
# 步骤3: 可视化
top_5_platforms.plot( kind='bar', figsize=(10, 6), rot=0)
plt.title('历史上总销量最高的Top 5游戏平台')
plt.xlabel('游戏平台')
plt.ylabel('全球总销量 (百万)')
plt.grid(axis='y', linestyle='--')
plt.show()
```

## <font color='red'> --- 【挑战2】从2000年到2016年，每年发行的游戏数量的变化趋势是怎样的？ --- </font>

```python
print("\n\n--- 【挑战2】2000-2016年每年发行游戏数量趋势 ---")
    
# 步骤1: 筛选出年份在2000年及之后的数据，并处理年份列中的缺失值
# 任务: 筛选出年份在2000年及之后的数据
games_after_2000 = df_games[(df_games['年份'] >= 2000) & (df_games['年份'] <= 2016)].dropna(subset=['年份'])
    
# 步骤2: 按 '年份' 分组，并使用 .size() 统计每年的游戏数量
yearly_game_counts = games_after_2000.groupby('年份').size()
    
print("2000年以来每年发行的游戏数量：\n", yearly_game_counts.head()) # 打印最后几行查看
    
# 步骤3: 绘制折线图
# 任务: 为 yearly_game_counts 绘制一个折线图
yearly_game_counts.plot(kind='line',figsize=(12, 7), marker='o')
plt.title('2000-2016年每年发行游戏数量趋势')
plt.xlabel('年份')
plt.ylabel('发行的游戏数量')
plt.grid(True)
plt.xticks(yearly_game_counts.index.astype(int)) # 确保X轴年份是整数
plt.show()
```

## <font color='red'> --- 【挑战3】欧洲和北美玩家最喜欢的游戏类型Top 3分别是什么？ --- </font>

```python
print("\n\n--- 【挑战3】欧洲与北美玩家最爱游戏类型Top 3对比 ---")
    
# 任务: 按 '类型' 分组, 同时计算 '北美销量(百万)' 和 '欧洲销量(百万)' 的总和
region_sales_by_genre = df_games.groupby('类型')[['北美销量(百万)','欧洲销量(百万)']].sum()
    
    # 任务: 分别对北美和欧洲的销量进行排序，并选出前3名
na_top_3_genres = region_sales_by_genre['北美销量(百万)'].sort_values(ascending=False).head(3)
eu_top_3_genres = region_sales_by_genre['欧洲销量(百万)'].sort_values(ascending=False).head(3)
    
print("【北美玩家】最喜欢的游戏类型Top 3:\n", na_top_3_genres)
print("\n【欧洲玩家】最喜欢的游戏类型Top 3:\n", eu_top_3_genres)
```
