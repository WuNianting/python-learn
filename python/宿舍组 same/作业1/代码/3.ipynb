```python
# -*- coding: utf-8 -*-
# 这是一个为学生准备的代码填空练习文件。
# 请在标记了 # TODO 的地方，填写正确的代码，以完成相应的分析任务。

# 导入我们需要的库，这是每次开始分析的“标准动作”
import pandas as pd
import matplotlib.pyplot as plt


# 设置matplotlib以正确显示中文，避免乱码问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 读取两个数据集
df_market = pd.read_csv('农产品销售数据.csv')
df_games = pd.read_csv('vgsales_cn.csv')

# 对农产品数据进行基础清洗
df_cleaned = df_market.drop(columns=['提取时间'])

print("--- 数据已准备就绪，请开始完成以下任务 ---")
```

# 任务一：农产品价格深度分析 

```python
print("\n--- 任务1.1: 精确筛选 ---")
# 需求: 筛选出“上海市江桥批发市场经营管理有限公司”中，“最高价”低于5元 并且 “最低价”高于2元的蔬菜。
task1_1 = df_cleaned[
    # TODO: 在下方括号内，填入三个筛选条件，并用 & 连接
    (df_cleaned['批发市场'] == '上海市江桥批发市场经营管理有限公司') &
    (df_cleaned['最高价'] < 5) &
    (df_cleaned['最低价'] > 2)
]
print("任务1.1 结果：\n", task1_1)
```

```python
print("\n--- 任务1.2: 多品类查询 ---")
# 需求: 使用 .isin() 方法，一次性找出所有“大白菜”、“西红柿”、“黄瓜”的交易数据。
# TODO: 在下面的列表中填入你想查询的蔬菜名称
target_produce = [ '大白菜', '西红柿', '黄瓜' ] 
task1_2 = df_cleaned[df_cleaned['品种'].isin(target_produce)]
print("任务1.2 结果：\n", task1_2.head())
```

```python
print("\n--- 任务1.3: 分布探索 ---")
# 需求: 计算一个新的列 `价差` (等于 最高价 - 最低价)，并为这个 `价差` 列画出直方图。
# TODO: 计算 '最高价' 和 '最低价' 的差值，赋值给新列 '价差'
# TODO: 选取 '价差' 这一列 (Series)，并调用 plot 方法绘制直方图
df_cleaned['价差'] = df_cleaned['最高价'] - df_cleaned['最低价']
df_cleaned['价差'].plot(kind='hist', bins=15, title='农产品价差分布直方图')
plt.xlabel('价差区间')
plt.ylabel('频数')
plt.show()
```

# 任务二：全球游戏市场洞察

```python
print("\n--- 任务2.1: 组合条件过滤 ---")
# 需求: 筛选出“任天堂(Nintendo)”发行的，或者 “动作(Action)”类型的，并且 全球销量(百万) 大于20的游戏。
task2_1 = df_games[
    # TODO: 在下方填入复杂的筛选条件，注意括号的使用，| 代表“或者”，& 代表“并且”
    ((df_games['发行商'] == 'Nintendo') | (df_games['类型'] == 'Action')) &
    (df_games['全球销量(百万)'] > 20)
]
print("任务2.1 结果：\n", task2_1)
```

```python
print("\n--- 任务2.2: 寻找关联性 ---")
```

```python
##  需求: 绘制一张散点图，探索“日本销量(百万)”和“全球销量(百万)”之间的关系。
- <>TODO: 在下方 plot 函数中，为 x 和 y 参数指定正确的列名
```

```python
df_games.plot(kind='scatter', x='日本销量(百万)', y='全球销量(百万)', title='日本与全球销量关系')
plt.grid(True)
plt.show()
```

```python
print("\n--- 任务2.3: (挑战任务)热门类型分析 ---")
# 需求: 找出全球总销量最高的Top 5游戏类型。
# TODO: 步骤1: 按 '类型' 分组，并计算 '全球销量(百万)' 的总和
genre_sales = df_games.groupby('类型')['全球销量(百万)'].sum()

# TODO: 步骤2: 对结果进行降序排序，并用 .head() 选取前5个
top_5_genres = genre_sales.sort_values(ascending= False).head()

print("销量最高的Top 5游戏类型：\n", top_5_genres)
top_5_genres.plot(kind='bar', title='Top 5 游戏类型全球总销量')
plt.ylabel('全球总销量 (百万)')
plt.show()

print("\n--- 所有练习已完成！ ---")
```
